```python
# ================= Fingerprint Authentication Template =================

# Stored fingerprints in database (simulated)
fingerprint_database = {
    "FP101": "Madhu",
    "FP102": "Ravi",
    "FP103": "Sita"
}

# Function to authenticate fingerprint
def authenticate_fingerprint(scanned_fingerprint):
    try:
        print("Scanning fingerprint...")

        if scanned_fingerprint in fingerprint_database:
            user = fingerprint_database[scanned_fingerprint]
            print("Fingerprint Found in Database")
            print(f"Welcome {user}")
            print("ACCESS GRANTED ✅")
        else:
            print("Fingerprint NOT Found in Database")
            print("ACCESS DENIED ❌")

    except Exception as e:
        print("System Error:", e)

# ===================== Test Scenarios =====================

# Case 1: Valid fingerprint
authenticate_fingerprint("FP101")

print("\n-------------------------\n")

# Case 2: Invalid fingerprint
authenticate_fingerprint("FP999")
```
